# 2.-

A continuaci√≥n est√° la programaci√≥n de la derivada, que hace exactamente lo que hace la definici√≥n de derivada. 

```python
def derivada(ec,variable,h=0.01):
    df = (ec(variable + h) - ec(variable))/h
    return df
```

Y la ecuaci√≥n para confirmar es:

```python
def ecuacion(x):
    return x**3 - 5*x + 2
```

Aqu√≠ se ve la gr√°fica de las tres ecuaciones, vemos que abajo la verde y la azul son indistinguibles, ya que la aroximaci√≥n es buena. 

```python
y = linspace(0,30,1000)
dg = derivada(ecuacion,y)
dg2 = 3*(y**2) - 5
g = y**3 - 5*y + 2

plot(y,dg2,y,dg,y,g)
```

Quitando la funci√≥n original, vemos que incluso para valores de h grandes, como 2 en este caso, la aproximaci√≥n es buena. 

```python
y = linspace(0,30,1000)
dg = derivada(ecuacion,y,h=2)
dg2 = 3*(y**2) - 5
g = y**3 - 5*y + 2

plot(y,dg2,y,dg)
```

# 3.-

El siguiente c√≥digo encuentra las ra√≠ces usando el m√©todo de Newton y la ecuaci√≥n programada en el inciso 2. Hace el m√©todo hasta que el punto evaluado sea menor que el error.

```python
def newton(ec,p,err=0.001):
    
    raiz = p
    while abs(ec(raiz)) > err:
        raiz = raiz - ec(raiz)/derivada(ec,raiz)
    
    return raiz
```

```python
def ec2(x):
    return sin(3*x) + 0.5*cos(2*x)*cos(2*x)
```

Aqu√≠ vemos que los puntos encontrados son efectivamente las ra√≠ces de la funci√≥n. Aqu√≠ se us√≥ el for para tomar muchos puntos iniciales diferentes y obtener las ra√≠ces.

```python
w = linspace(0,30,1000)
for i in arange(600):
    raiz = newton(ec2,0.1*i)
    plot(raiz,0,'o')

plot(w,ec2(w))
xlim(0,30)
grid(True)
show()
```

# 4.-

Primero tomaremos como ec3 bessel de orden cero y ec4 su derivada.

```python
def ec3(x):
    return jv(0,x)

def ec4(x):
    return -jv(1,x)
```

Vemos que tomando la derivada num√©rica programada y el m√©todo de Newton anterior se encuentran las ra√≠ces.

```python
ww = linspace(0,10,200)
plot(ww,jv(0,ww))
xlim(0,10)
grid(True)

for i in arange(600):
    raiz = newton(ec3,0.1*i)
    plot(raiz,0,'o')
```

Programamos un Newton especial donde se toma la derivada anal√≠tica.

```python
def newton_bessel(ec,p,err=0.001):
    
    raiz = p
    while abs(ec(raiz)) > err:
        raiz = raiz - ec(raiz)/ec4(raiz)
    
    return raiz
```

Vemos que aqu√≠ tambi√©n se encuentran as ra√≠ces correctas.

```python
ww = linspace(0,10,200)
plot(ww,jv(0,ww))
xlim(0,10)
grid(True)

for i in arange(600):
    raiz = newton_bessel(ec3,0.1*i)
    plot(raiz,0,'o')
```

Vemos que ambos m√©todos dan ra√≠ces muy similares. En este caso la primer ra√≠z de manera con derivada num√©rica o anal√≠tica difieren en el cuarto decimal. 

Num√©rica:

```python
newton(ec3,1)
```

```python
jv(0,2.4046317112393911)
```

Anal√≠tica:

```python
newton_bessel(ec3,1)
```

```python
jv(0,2.4045555512049832)
```

# 5.-

Sea una matriz

$$
\left( \begin{array}{ccc}
a & b \\
c & d \end{array} \right)
$$

que veremos aqu√≠ como:

[[a,b],[c,d]]

Sabemos que si el determinante es cero la matriz no tiene inversa, para lo cual se usa el primer if de la funci√≥n. Si no, voltearemos los de la diagonal y les cambiaremos el signo a los otros.

```python
def inversa(matriz):
    
    if matriz[0][0]*matriz[1][1] - matriz[0][1]*matriz[1][0] == 0:
        print "No tiene matriz inversa"
        
    else:
        det_i = 1/(matriz[0][0]*matriz[1][1] - matriz[0][1]*matriz[1][0])
        a = matriz[1][1]*det_i
        b = -matriz[0][1]*det_i
        c = -matriz[1][0]*det_i
        d = matriz[0][0]*det_i
        
        print [[a,b],[c,d]]
```

Vemos del siguiente ejemplo que efectivamente nos da la matriz inversa. Es importante escribir los n√∫meros con un punto porque si no no hace bien la divisi√≥n del determinante.

```python
mat = [[1.,2.],[3.,4.]]
inversa(mat)
```

# 6.-

Queremos primero convertir nuestro n√∫mero en un entero, para eso si no es divisible entre uno lo multiplicamos por diez hasta que lo sea. 

Luego le restamos uno a nuestro n√∫mero hasta que sea divisible entre diez y cada paso que se hace lo vamos sumando a una "n". Cuando es divisible entre diez lo dividimos entre diez y repetimos. As√≠ se suma cada d√≠gito.

```python
def sumador(num):
    n = 0
    while num%1 != 0:
        num = num*10
        
    while num >= 1:
        while num%10 != 0:
            num = num - 1
            n += 1
        num = num*0.1
    
    return n
```
